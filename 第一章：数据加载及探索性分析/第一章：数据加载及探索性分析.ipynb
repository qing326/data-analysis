##### 1.1.4.2 DataFrame创建
- 列表创建（逐行输入）
- 字典嵌套列表（逐列输入）
- 列表嵌套字典(先行后列，逐个输入）

```python
#列表创建（逐行输入）
data = [['Tom', 28],
        ['Jack',34],['Steve',29],['Ricky',42]]
df = pd.DataFrame(data=data,
                  index=['rank%d'%i for i in range(4)],# index默认为range(n)
                  columns=['Name','Age'])
df
```

### 1.5 筛选
- df[ ]（筛选组合条件）
- df.loc[行标签，列标签]或df.iloc[行位置，列位置]
- df[ ].str.contains( ) 适合字符串

```python
#以"Age"为筛选条件，显示年龄在10岁以下的乘客信息。
df[df["Age"]<10].head()
```

```python
#以"Age"为条件，筛选年龄在10岁以上和50岁以下的乘客信息，并将数据命名为midage
midage=df[(df["Age"]>10)&(df['Age']<50)]
midage.head()
```

```python
#将midage的数据中第100行的"Pclass"和"Sex"的数据显示出来
midage = midage.reset_index(drop=True)#使用reset_index()将midage数据索引重置为连续索引
midage.head()
midage.loc[100, ['Pclass','Sex']]
```

```python
#使用使用loc将midage筛选出中第100，105，108行的"Pclass"，"Name"和"Sex"的数据
midage.loc[[100,105,108], ['Pclass','Name','Sex']]
```

```python
#使用使用iloc将midage筛选出中第100，105，108行的"Pclass"，"Name"和"Sex"的数据
midage.iloc[[100,105,108],[2,3,4]]
```

**对比`iloc`和`loc`的异同**
- df.loc[ ]通过标签索引，而df.iloc[ ]通过位置索引

#### 1.6.1 利用Pandas对示例数据排序
- sort_values( )值排序
- sort_index( )索引排序
- 默认参数 ascending=True 为升序

```python
#自行构建DataFrame数据，并根据某列进行升序排列
array=np.arange(8).reshape((2,4))#生成一个二维数组（2*4）,第一列：0，1，2，3 第二列：4，5，6，7
data=pd.DataFrame(data=array,
                 index=[2,1],
                 columns=['a','b','c','d'])
data.sort_values('a',ascending=True)#根据a列值升序
```

**【总结】**按索引值进行排序
- df.sort_index(axis=0,ascending=True)
- axis=0(行);axis=1(列)
- ascending=True(升序)

```python
data.sort_index(axis=0,ascending=True).head() #让行索引升序
data.sort_index(axis=1,ascending=True).head() #让列索引升序
data.sort_index(axis=1,ascending=False).head() #让列索引降序排序
data.sort_values(['a','b'],ascending=False).head() #让任选两列数据同时降序排序
```
